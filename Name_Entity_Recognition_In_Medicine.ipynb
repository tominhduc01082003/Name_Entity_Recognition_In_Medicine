### TODO_1: Load datasets

```python
from datasets import load_dataset
import os

DATA_DIR = "Datasets"   

data_files = {
    "train": os.path.join(DATA_DIR, "viet_med-ner-train.arrow"),
    "validation": os.path.join(DATA_DIR, "viet_med-ner-validation.arrow"),
    "test": os.path.join(DATA_DIR, "viet_med-ner-test.arrow")
}

dataset = load_dataset("arrow", data_files=data_files)
print(dataset)
```

```python
# dataset = dataset.remove_columns(["audio", "duration"])
print("Columns:", dataset["train"].column_names)
print("Sample 0 (train):")
print(dataset["train"][0])
```

```python
all_labels=set()
for example in dataset["train"]:
    all_labels.update(example["labels"])
print(sorted(all_labels))
print(len(all_labels))
    
```
